```python
import yfinance as yf
import datetime
data = yf.download("AAPL", period="1mo")
print(data)
```

```python
def get_historical_data_yfinance(ticker, start_date = "2000-1-1", end_date = datetime.date.today()):

    prices = []

    try:
        yf_data = yf.download(ticker,start_date, end_date, interval="1d")
        prices = list(yf_data.to_records())
    except Exception as e:
        print(f"Could not download the price data due to the following error:{e}")

    return prices

ticker = 'MSFT'
prices = get_historical_data_yfinance(ticker)
print(prices)
```
